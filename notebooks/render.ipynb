```python
# Chromedriverã®ãƒ‘ã‚¹
chromedriver_path = 'C:/Users/varyu/Desktop/development/twitter/.venv/Lib/site-packages/chromedriver_binary/chromedriver.exe'
```

```python
# WebDriverã®Serviceã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆã‚’ä½œæˆ
service = Service(chromedriver_path)

# WebDriverè¨­å®š
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # ãƒ˜ãƒƒãƒ‰ãƒ¬ã‚¹ãƒ¢ãƒ¼ãƒ‰ã§ãƒ–ãƒ©ã‚¦ã‚¶ã‚’èµ·å‹•
options.add_argument("window-size=1200x600") # ãƒ“ãƒ¥ãƒ¼ãƒãƒ¼ãƒˆã‚µã‚¤ã‚ºã‚’èª¿æ•´

driver = webdriver.Chrome(service=service, options=options)
```

```python
# HTMLã‚’ãƒ¬ãƒ³ãƒ€ãƒªãƒ³ã‚°
driver.get("data:text/html;charset=utf-8," + html_content)

# Twitterã‚¦ã‚£ã‚¸ã‚§ãƒƒãƒˆãŒãƒ­ãƒ¼ãƒ‰ã•ã‚Œã‚‹ã¾ã§å¾…ã¤
wait = WebDriverWait(driver, 20)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ".twitter-tweet")))

# ã‚¹ã‚¯ãƒªãƒ¼ãƒ³ã‚·ãƒ§ãƒƒãƒˆå–å¾—
screenshot = driver.get_screenshot_as_png()
driver.quit()
```
